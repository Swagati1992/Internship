```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")
```

```python
url="https://www.cnbc.com/world/?region=world"
```

```python
page=requests.get(url)
page.status_code
```

```python
soup=BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_headlines=soup.find_all("a",class_="LatestNews-headline")
scraped_headlines
```

```python
headlines=[]
for i in scraped_headlines:
    headlines.append(i.get_text().replace("\n"," ").strip())
print(headlines)
```

```python
scraped_time=soup.find_all("time",class_="LatestNews-timestamp")
scraped_time
```

```python
time=[]
for i in scraped_time:
    time.append(i.get_text().replace("\n"," ").strip(" "))
print(time)
```

```python
scraped_links=soup.find_all("a",class_="LatestNews-headline")
scraped_links
```

```python
links=[]
for i in scraped_links:
    links.append(i.get("href").replace("\n"," ").strip(" "))
print(links)
```

```python
df=pd.DataFrame()
df['Headlines']=headlines
df['Time']=time
df['Links']=links
df.head()
```
